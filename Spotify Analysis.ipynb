# Data Analysis

```python
# Top 5 popular artists
```

```python
top_5_artist = df.groupby('artist').count().sort_values(by='song_title',ascending=False)['song_title'][:5]
top_5_artist
```

```python
top_5_artist.plot.barh()
plt.show()
```

```python
# Top 5 loudest tracks
```

```python
top_5_loudest_tracks = df[['loudness','song_title']].sort_values(by='loudness',ascending=True)[:5]
top_5_loudest_tracks
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='loudness',y='song_title',data=top_5_loudest_tracks)
plt.title('Top 5 Loudest Tracks')
plt.show()
```

```python
# Artist with most danceability song
```

```python
df.head(1)
```

```python
top_5_artist_danceability_songs = df[['danceability','song_title','artist']].sort_values(by='danceability'
                                                                               ,ascending=False)[:5]
top_5_artist_danceability_songs
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='danceability',y='artist',data=top_5_artist_danceability_songs)
plt.title('Artists with Danceability Songs')
plt.show()
```

```python
 # Top 10 instrumentalness songs
```

```python
top_10_instrumentalness_songs = df[['instrumentalness','song_title','artist']].sort_values(by='instrumentalness',
                                                                                          ascending=False)[:5]
top_10_instrumentalness_songs
```

```python
plt.figure(figsize=(12,7))
plt.pie(x='instrumentalness',data=top_10_instrumentalness_songs,autopct='%1.2f%%',labels=top_10_instrumentalness_songs.song_title)
plt.title('top_10_instrumentalness_songs')
plt.show()
```

# Multiple feature plot

```python
interest_feature_cols = ['tempo','loudness','acousticness','danceability','duration_ms','energy',
                        'instrumentalness','liveness','speechiness','valence']
```

```python
for feature_col in interest_feature_cols:
    pos_data = df [df["target"] == 1][feature_col]
    neg_data = df[df["target"] == 0][feature_col]
    
    plt. figure(figsize= (12, 7))
    sns.distplot (pos_data, bins=30, label="Positive", color="green" ) 
    sns.distplot (neg_data, bins=30, label="Negative", color="red")
    
    plt. legend (loc="upper right")
    plt.title(f"Positive And Negative Histogram Plot For {feature_col}")
    plt.show ()
```
